```python
import pandas as pd

# --- 1) Read with correct separators ---
product = pd.read_csv("tianchi_product.csv",
                      sep="\x01", header=0,
                      names=["item_id","title","pic_url","cat_id","merchant_id","brand_id"],
                      dtype="string")

log_all = pd.read_csv("tianchi_log_ABC_merged_cleaned.csv",
                      sep=",", header=0, 
                      dtype="string")

# --- 2) Standardise the join key ---
product["item_id_clean"] = product["item_id"].str.extract(r"(\d+)", expand=False)
log_all["item_id_clean"]  = log_all["item_id"].str.extract(r"(\d+)", expand=False)

# --- 3) Drop missing keys & de-duplicate per key (keep first non-null values) ---
product = (product.dropna(subset=["item_id_clean"])
                   .sort_values("item_id_clean")
                   .drop_duplicates(subset=["item_id_clean"], keep="first"))

log_all = (log_all.dropna(subset=["item_id_clean"])
                  .sort_values(["user_id","time_col"])  # optional stable order
                  .reset_index(drop=True))

# --- 4) Merge (ALL columns kept; rename clashes with suffixes) ---
keep_prod = product.drop(columns=["item_id"])   # keep only metadata + item_id_clean
merged = log_all.merge(
    keep_prod, on="item_id_clean", how="inner",
    indicator=False,           # <- no _merge column
    validate="many_to_one"     # optional sanity check
)


print("Merged shape:", merged.shape)
print("Unique items in merged:", merged["item_id_clean"].nunique())


# --- 5) Save ---
merged.to_csv("tianchi_log_product_merged3.csv", index=False, encoding="utf-8")
print("Saved: tianchi_log_product_merged3.csv")
```
